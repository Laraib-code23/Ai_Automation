```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
dict={"marks":['85','45','72','60','30'],
      "Result":["pass","fail","pass","pass","fail"]
     }
```

```python
df=pd.DataFrame(dict)
df
```

```python
df.to_csv('Resul.csv') #To Creat csv file
```

```python
df=pd.read_csv('Resul.csv')
df
```

```python
x=df[['marks']]
x
```

```python
y=df[['Result']]
y
```

```python
y = df['Result']
y_encoded = y.map({'pass': 1 , 'fail': 0})
y_encoded
```

```python
model=LogisticRegression()
model.fit(x,y_encoded)
```

```python
test_marks = [[45],[90],[58]]
prediction = model.predict(test_marks)
```
