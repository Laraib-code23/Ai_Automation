```python
import pandas as pd
import numpy as np
import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
```

```python
df = pd.read_csv("/content/mental_health_dataset.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['mental_health_risk_Encoder'] = le.fit_transform(df['mental_health_risk'])
df['gender_Encoder']= le.fit_transform(df['gender'])
df['mental_health_history_Encoder'] = le.fit_transform(df['mental_health_history'])
df['seeks_treatment_Encoder'] = le.fit_transform(df['seeks_treatment'])
df.head()
```

```python
df.drop(columns=['mental_health_risk','gender','work_environment','mental_health_history','seeks_treatment','employment_status'],inplace=True)
df.head()
```

```python
X=df.drop(columns =['mental_health_risk_Encoder'])
X.head()
```

```python
y=df['mental_health_risk_Encoder']
y
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = model.predict(X_train)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = model.predict(X_test)
# Assuming y_true and y_pred (predicted classes from probabilities)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
y_pred_probs = model.predict_proba(X_test)
log_loss_value = log_loss(y_test, y_pred_probs)
print("Logg Loss value:", log_loss_value)
```

```python
fig = px.bar(df,x='gender_Encoder', y='anxiety_score',
             title='Average Anxiety by Gender',
             color='gender_Encoder',
             barmode='group',
             height=400)
fig.show()
```

```python
corr_matrix = df[['stress_level', 'sleep_hours', 'physical_activity_days',
                  'depression_score', 'anxiety_score',
                  'social_support_score', 'productivity_score']].corr()
fig = ff.create_annotated_heatmap(
    z=corr_matrix.values,
    x=list(corr_matrix.columns),
    y=list(corr_matrix.index),
    annotation_text=corr_matrix.round(2).values,
    colorscale='Viridis'
)
fig.update_layout(title='Correlation Between Mental Health Indicators')
fig.show()
```
